# Package `nelpy.plotting` examples

Here we will look at some of the different plot types and utilities in the `nelpy.plotting` module.

Let's start with the required imports.

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

import nelpy as nel  # recommended import for nelpy
import nelpy.plotting as npl  # recommended import for the nelpy plotting library

%matplotlib inline
```

### Powerful scalebar implementation

Nelpy includes a powerful and flexible scalebar implementation based on an excellent gist by Dan Meliza (see https://gist.github.com/dmeliza/3251476). Nelpy extends the basic functionality in a number of ways, and formats the scalebar labels correctly.

```python
fig = plt.figure(1, figsize=(5, 3))
ax = fig.add_subplot(111)

ax.yaxis.set_major_formatter(npl.FixedOrderFormatter(-2))
npl.add_scalebar(ax=ax, hidey=False, labelx='xxx', labely='2 cm', sep=15, sizex=0.3, sizey=0.4, pad=0.5, lw=5, ec='deepskyblue', fc='green', fontsize=16)
```

```python
fig = plt.figure(1, figsize=(5, 3))
ax = fig.add_subplot(111)

ax.yaxis.set_major_formatter(npl.FixedOrderFormatter(-2))
npl.add_scalebar(ax=ax, hidey=False, labelx='xxx', labely='2 cm', sep=15, sizex=0, sizey=0.4, pad=0.5, lw=5, ec='deepskyblue', fc='green', fontsize=16)
```

```python
fig = plt.figure(1, figsize=(5, 3))
ax = fig.add_subplot(111)

ax.yaxis.set_major_formatter(npl.FixedOrderFormatter(-2))
npl.add_scalebar(ax=ax, hidey=False, labelx='xxx', labely='2 cm', sep=15, sizex=0.3, sizey=0, pad=0.5, lw=5, ec='deepskyblue', fc='green', fontsize=16)
```

```python
fig = plt.figure(1, figsize=(5, 3))
ax = fig.add_subplot(111)

ax.yaxis.set_major_formatter(npl.FixedOrderFormatter(-2))
npl.add_scalebar(ax=ax, hidey=False, hidex=False, sizex=0.2, sizey=0.2)
```

```python
fig = plt.figure(1, figsize=(5, 3))
ax = fig.add_subplot(111)

ax.yaxis.set_major_formatter(npl.FixedOrderFormatter(-2))
npl.add_scalebar(ax=ax, sizex=0.2, sizey=0.2)
```

```python
fig = plt.figure(1, figsize=(5, 3))
ax = fig.add_subplot(111)

ax.yaxis.set_major_formatter(npl.FixedOrderFormatter(-2))
npl.add_scalebar(ax=ax, labelx='f')
```
